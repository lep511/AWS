## Configuración e implementación de Amazon VPC con varias subredes

### ¿Qué es una Amazon VPC?

Una VPC es una nube privada virtual. En esencia, una VPC es un centro de datos virtual en la nube. Las VPC de Amazon son redes virtuales, asociadas a una única región de AWS, y es un servicio que define un límite en torno a los servicios y recursos de AWS que los clientes deciden implementar y el modo en que dichos servicios y recursos se comunican entre sí y con redes externas como Internet. AWS admite configuraciones de nube híbrida que facilitan la conexión entre una Amazon VPC y una ubicación local, como un centro de datos físico.  

Existen dos tipos de Amazon VPC en una cuenta de AWS: una Amazon VPC predeterminada (por defecto) y una Amazon VPC personalizada.

#### Amazon VPC por defecto

Al crear una cuenta de AWS, se crean Amazon VPC predeterminadas en cada región de AWS compatible. Al utilizar la Amazon VPC predeterminada, puede comenzar inmediatamente a implementar recursos y no tener que pensar en la red subyacente.  
viñeta
Cada Amazon VPC predeterminada crea una subred pública dentro de cada zona de disponibilidad dentro de la región soportada. 

* Cada subred pública se configura con una ruta predeterminada para todo el tráfico entrante y saliente que dirige el tráfico IP a Internet general. 
* AWS establece la configuración que permite todo el tráfico, por lo que no hay privacidad y aislamiento por defecto.
* Sólo se permite una Amazon VPC por defecto por Región.
* Cada valor predeterminado viene con un rango de Amazon VPC Classless Inter-Domain Routing (CIDR), que es un rango determinado de direcciones IP.

Este CIDR de Amazon VPC predeterminado define el rango inicial y final de la dirección IP que puede utilizar la Amazon VPC predeterminada. Todo lo que se encuentra dentro de una Amazon VPC utiliza este rango CIDR. Todas las comunicaciones a la Amazon VPC tendrán que utilizar el CIDR de Amazon VPC y las comunicaciones salientes se realizarán desde este CIDR de Amazon VPC.  

Todas las Amazon VPC predeterminadas se configuran del mismo modo. Para una mayor resistencia, la Amazon VPC predeterminada se divide automáticamente en subredes a través de las zonas de disponibilidad.  

* Cada Amazon VPC predeterminada está configurada para tener una subred ubicada en cada zona de disponibilidad de esa región.  
* Cada subred de la Amazon VPC predeterminada utiliza parte del rango de direcciones IP del CIDR de la Amazon VPC. 
* El intervalo de direcciones IP de cada subred debe ser único con respecto al intervalo de direcciones IP de las demás subredes y no puede solaparse.  

Si falla una zona de disponibilidad en su Amazon VPC predeterminada, la subred asociada también fallará, pero con la Amazon VPC predeterminada, las demás subredes de otras zonas de disponibilidad siguen funcionando. 

El CIDR IPv6 de Amazon VPC para su rango de subred es /64, y el rango CIDR de Amazon VPC es /56. La ruta local se utiliza para comunicarse entre subredes dentro de su Amazon VPC. El CIDR IPv4 predeterminado de Amazon VPC, 172.31.0.0/16, es siempre el mismo, y también está diseñado y configurado de la misma forma.

#### Amazon VPC personalizada

Una Amazon VPC personalizada es una red virtual aislada lógicamente dentro de una Región soportada bajo una única cuenta, lo que la convierte en un servicio regional. 

* A diferencia de una Amazon VPC predeterminada, cada componente de una Amazon VPC personalizada debe definirse explícitamente al crearla; no se permite la entrada ni la salida de nada sin una configuración explícita. 
* Algunas decisiones, como la compatibilidad con IPv4 e IPv6 y el bloque CIDR para la Amazon VPC, no pueden modificarse posteriormente. 
* Otras características de una Amazon VPC, como las subredes, el enrutamiento y los puntos finales de la VPC, pueden modificarse según sea necesario.

Una Amazon VPC personalizada, similar a una Amazon VPC predeterminada, proporciona una red virtual aislada lógicamente que soporta el despliegue de recursos y servicios soportados por la región en la que se crea la Amazon VPC. Se proporciona un control completo sobre la red virtual definida por la Amazon VPC. 

Esto incluye

* Elegir los rangos de direcciones IP soportados definiendo cada subred
* Administración del flujo de tráfico interno de la red 
* Administración de cómo el tráfico entra y sale de la Amazon VPC a través de tablas de rutas y puertas de enlace de red

En una sección posterior de este curso, aprenderá cómo se puede configurar una subred pública para permitir que una aplicación o recurso tenga acceso a Internet y cómo se puede utilizar una subred privada, una subred sin acceso a la red externa, para ayudar a proteger una base de datos o un sistema backend. También aprenderá sobre las múltiples capas de seguridad disponibles a través de una Amazon VPC y cómo una Amazon VPC puede utilizar, por ejemplo, ACL de red y grupos de seguridad para controlar a quién y a qué se le permite el acceso a los recursos desplegados dentro de la Amazon VPC.

AWS admite la ampliación de su entorno de AWS mediante el establecimiento de una conexión segura entre una Amazon VPC y una red local utilizando:

* AWS Direct Connect
* Red privada virtual (VPN) de sitio a sitio de AWS
* VPN de cliente de AWS

Esto se trata con más detalle en el curso Opciones de conectividad de red de AWS.

### En resumen

**VPC Predeterminada:**

* Cada cuenta de AWS viene con una Amazon VPC predeterminada, lo que permite al cliente comenzar inmediatamente a crear un entorno en la nube de AWS. 

* La Amazon VPC predeterminada se configura inicialmente con un bloque CIDR IPv4 /16, una subred predeterminada /20 para las zonas de disponibilidad de las regiones, una puerta de enlace a Internet, una ruta predeterminada a la puerta de enlace a Internet, un grupo de seguridad predeterminado, una ACL de red predeterminada y un conjunto de opciones predeterminadas del protocolo de configuración dinámica de host (DHCP). 

* El enrutamiento predeterminado a la puerta de enlace de Internet garantiza que todas las subredes iniciales dentro de la Amazon VPC predeterminada tengan una ruta de salida a Internet y estén configuradas para asignar cualquier instancia de Amazon Elastic Compute Cloud (Amazon EC2) lanzada dentro de la Amazon VPC predeterminada con direcciones IP públicas y privadas.

* La siguiente imagen ilustra todas las características de una Amazon VPC predeterminada. Por ejemplo, observe que el bloque CIDR para una Amazon VPC predeterminada es siempre 172.31.0.0/16 esto proporciona 65.536 direcciones IP.

* AWS crea una subred predeterminada en cada zona de disponibilidad de la región. Cada subred predeterminada tiene un bloque CIDR de tamaño /20, lo que permite hasta 4.096 direcciones IP.

* AWS crea una ACL predeterminada. La ACL predeterminada permite todo el tráfico entrante y saliente.

* También se crea una tabla de rutas predeterminada. La tabla de rutas predeterminada tiene una ruta que permite todo el tráfico entrante y saliente a través de la puerta de enlace a Internet.

**VPC Personalizada:**

* En cada cuenta de AWS, también puede crear una Amazon VPC personalizada, que es una red aislada, a diferencia de la Amazon VPC predeterminada. Con una Amazon VPC personalizada, no se permite la entrada ni la salida de nada sin una configuración explícita. 

* Las VPC personalizadas son completamente configurables, a diferencia de la Amazon VPC predeterminada que AWS configura cuando se crea la cuenta de AWS.

* Otra diferencia entre las VPC personalizadas y las predeterminadas es que con una Amazon VPC personalizada se puede elegir entre tenencia predeterminada o tenencia dedicada. La tenencia predeterminada se utiliza para los recursos aprovisionados dentro de la Amazon VPC que se aprovisionan en hardware compartido con otros. La tenencia dedicada es el hardware que está dedicado exclusivamente a usted. No se comparte con otros, pero si se elige la tenencia dedicada, no podrá modificarse posteriormente. Será necesario eliminar la VPC de Amazon y volver a configurarla para elegir la tenencia predeterminada. La tenencia dedicada cuesta más, por lo que, para optimizar los costes, elija la tenencia predeterminada a menos que exista un motivo específico para elegir la tenencia dedicada. 

* Las VPC personalizadas también pueden utilizar bloques CIDR IPv4 e IP públicas. AWS asigna un bloque CIDR IPv4 privado cuando se aprovisiona la  VPC personalizada. Se puede elegir entre el rango CIDR más grande con un /16 que tiene 65.536 direcciones IP o el más pequeño /28 que da 16 direcciones IP. También con una Amazon VPC personalizada, existe un bloque secundario opcional de direcciones IPv4. Las Amazon VPC personalizadas también se pueden configurar con un único bloque CIDR /56 IPv6 asignado; AWS asigna automáticamente el CIDR IPv6, a diferencia de lo que ocurre con IPv4.

* Las VPC personalizadas también disponen de un sistema de nombres de dominio (DNS) totalmente aprovisionado que es la dirección IP de red +2.

**Consideraciones adicionales sobre Amazon VPC:**

* La única diferencia arquitectónica entre una subred pública y privada es que una subred pública tiene una ruta a una puerta de enlace de Internet.

* Por defecto, Amazon VPC gestiona el DNS. Sin embargo, es posible utilizar Amazon Route 53 para crear su propio DNS dentro de una Amazon VPC con zonas hospedadas privadas.

* Todo el tráfico es unicast y Amazon VPC no requiere el protocolo de resolución de direcciones (ARP).

* Por defecto, todas las subredes de una Amazon VPC se pueden comunicar entre sí. Puede utilizar el ACL de red para restringir el tráfico que entra y sale de sus subredes.

* Todo el tráfico entre dos puntos de la misma VPC se reenvía directamente.

### Seguridad

Al crear Amazon VPC, es fundamental integrar la seguridad en todas las capas. En esta sección y en la siguiente, revisará tres controles de acceso: AWS Identity and Access Management (IAM), las listas de control de acceso a la red (ACL de red) y los grupos de seguridad.

#### Visión general de IAM

IAM le permite controlar quién puede configurar y administrar sus Amazon VPC. Le permite crear y asignar políticas de permisos basadas en grupos, usuarios y roles. Estos permisos especifican lo que cada entidad puede y no puede hacer con los recursos que especifique.

Cuando se crea una cuenta de AWS, AWS crea un usuario raíz de la cuenta que tiene todos los permisos para la cuenta de AWS, y los permisos para el usuario raíz de la cuenta no se pueden ajustar. Sin embargo, en la mayoría de los casos es necesario conceder permisos a otras personas de tu organización dentro de la cuenta de AWS. 

Las organizaciones tienen diferentes usuarios, tal vez un equipo de almacenamiento, ingenieros, administradores de sistemas, y esas personas necesitan acceso a la cuenta de AWS, y también necesitan acceso a los servicios en la cuenta de AWS. La mejor práctica de AWS es seguir el principio de mínimo privilegio, lo que significa dar acceso a los usuarios de la cuenta de AWS únicamente a los servicios específicos a los que necesitan acceder. Los usuarios en una cuenta de AWS sólo necesitan permisos para hacer su trabajo; los permisos adicionales aumentan el riesgo de acciones que están fuera del alcance de su función individual y pueden tener impactos imprevistos en su entorno de AWS.  

Al utilizar el principio del mínimo privilegio, se limita el riesgo potencial y se garantiza que los usuarios sólo tienen los permisos necesarios, nada más. Cada cuenta de AWS tiene una cuenta de usuario raíz. El usuario raíz de la cuenta siempre tiene acceso completo a la cuenta y a los servicios de AWS, lo que le permite anular cualquier control de seguridad excesivamente restrictivo promulgado por cuentas IAM no raíz.

* Los usuarios IAM representan personas y también aplicaciones que necesitan acceso a una cuenta AWS. 

* Los grupos de IAM son grupos de usuarios relacionados, por ejemplo, su equipo de desarrollo, administradores de sistemas, ingenieros de almacenamiento y el equipo de finanzas.

* Los roles IAM son utilizados por los servicios de AWS. Los roles IAM también se pueden utilizar para conceder acceso externo a su cuenta de AWS junto con el acceso a los recursos y servicios de la cuenta de AWS. Por ejemplo, una instancia de Amazon Elastic Compute Cloud (Amazon EC2) dentro de su cuenta de AWS requiere acceso programable a Amazon CloudWatch, Amazon Simple Storage Service (Amazon S3), y similares.  

* Las políticas de IAM se utilizan para permitir o denegar el acceso a los servicios de AWS. Las políticas IAM deben adjuntarse a un usuario IAM, un grupo IAM o un rol IAM. AWS proporciona políticas preconfiguradas, AWS Managed Policies, que se pueden asignar según sea necesario. Los clientes también pueden crear políticas personalizadas en línea que permitan o denieguen combinaciones únicas de permisos que mejor se adapten al entorno de AWS del cliente. Por sí solas, las políticas se quedan ahí, para actuar deben estar asociadas a un usuario, un grupo o un rol.  

* En una visión general de alto nivel, IAM actúa como un proveedor de identidad (IdP), y gestiona las identidades dentro de una cuenta de AWS. IAM autentica estas identidades facilitando las actividades de inicio de sesión de la cuenta de AWS para que se les permita iniciar sesión en la cuenta de AWS, y luego autoriza a esas identidades a acceder a los recursos o denegar el acceso a los recursos sobre la base de las políticas adjuntas.

#### ACL de red

Las listas de control de acceso a la red (ACL de red) son un tipo de filtro de seguridad similar a un firewall que puede filtrar el tráfico que intenta entrar o salir de una subred. Las ACL de red se adjuntan a nivel de subred y, por defecto, se crea una ACL de red para una Amazon VPC predeterminada y se asocia a todas las subredes de su Amazon VPC predeterminada. 

Las ACL de red administran el tráfico que entra o sale de una subred porque las ACL de red están asociadas a la subred, no a los recursos dentro de la subred. Las ACL de red solo administran el tráfico que cruza el límite de la subred.

Si dos instancias de Amazon EC2 de una subred se están comunicando, las ACL de red no tendrán ninguna implicación si la comunicación entre las dos instancias no cruza el límite de la subred.

![image](images/imag02.png)

Por defecto, todas las Amazon VPC vienen con una lista de control de acceso a la red (ACL de red) predeterminada, asociada automáticamente a cualquier subred que no esté asociada a otra ACL de red. Por defecto, la ACL de red predeterminada permite la entrada y salida de todo el tráfico. La ACL de red por defecto sólo se aplica a las subredes a las que está asociada.

Al igual que con las tablas de rutas, puede editar la ACL de red predeterminada o crear nuevas ACL de red que pueden asociarse a una o más subredes.  

Las ACLs de red son una buena forma de limitar el acceso de un amplio rango de direcciones IP a o desde una subred. Por ejemplo, si estás protegiendo una aplicación empresarial y quieres impedir que entre todo el tráfico público, deberías configurar la ACL de red para que sólo permita el paso de direcciones IP privadas, proporcionando así una línea de defensa adicional.

#### Grupos de seguridad

Los grupos de seguridad son la otra característica de seguridad de una Amazon VPC. A diferencia de las ACL de red, los grupos de seguridad se adjuntan a los recursos de AWS, concretamente a las interfaces de red elásticas (ENI), no a las subredes de Amazon VPC. 

Los grupos de seguridad ofrecen algunas ventajas en comparación con las ACL de red, ya que pueden reconocer los recursos de AWS y filtrar en función de ellos. Los grupos de seguridad pueden hacer referencia a otros grupos de seguridad y también a sí mismos.  

Sin embargo, los grupos de seguridad no son capaces de bloquear tráfico explícitamente. Si necesita bloquear una dirección IP determinada o un bloque de direcciones IP, necesitará la ayuda de las ACL de red.

![image](images/imag03.png)

Las instancias de Amazon EC2 disponen de una capa adicional de seguridad de tráfico a través de los grupos de seguridad. Estos operan en el límite de la instancia.

Nota: Aunque las instancias pueden pertenecer al mismo grupo de seguridad (como se muestra aquí), eso no significa que las instancias del mismo grupo puedan acceder entre sí de forma predeterminada. Los grupos de seguridad son sólo grupos de reglas que se aplican a cada instancia por separado.

Las ACL de red especifican explícitamente qué tráfico está o no está permitido. Las reglas de los grupos de seguridad sólo especifican qué tráfico está permitido, mientras que el resto del tráfico está bloqueado.

Por defecto, los nuevos grupos de seguridad no tienen reglas de tráfico entrante, por lo que bloquean todo el tráfico entrante. Esto evita exponer accidentalmente su nueva instancia a Internet sin los controles de seguridad adecuados. 

Antes de cambiar las reglas de los grupos de seguridad de sus instancias, asegúrese de restringirlas correctamente. La seguridad dentro de su Amazon VPC es su responsabilidad.

Controles sin estado comparados con los controles con estado

Las ACL de red son sin estado, lo que significa que si se permite la entrada de tráfico, la respuesta de salida a ese tráfico NO se permite por defecto. 

* Para las reglas ACL de red, será necesario añadir explícitamente la dirección y el puerto de entrada y salida. 

* Las ACL de red sólo ven el tráfico que va en una dirección, por lo que si hay un permiso para una regla de entrada, también debe haber un permiso para la regla de salida. Entonces la ACL de red verá explícitamente que el tráfico que fue permitido en la entrada también está permitido en la salida. 

* Las ACLs de red ven el tráfico como dos flujos diferentes, se necesitan dos reglas, una regla para cada flujo. Si no se añade una regla de salida, el tráfico sólo se permitirá de entrada.

Los grupos de seguridad, sin embargo, son de estado. Si se permite la entrada de tráfico, la respuesta de salida a ese tráfico se permite automáticamente. 

* Los grupos de seguridad tienen reglas de entrada y salida, pero como los grupos de seguridad son de estado, eso significa que si se permite la entrada de tráfico, se permite automáticamente la salida de ese tráfico.  

* Los grupos de seguridad ven el tráfico entrante y saliente como parte del mismo flujo.  

* Una diferencia entre los grupos de seguridad y las ACL de red es que los grupos de seguridad reconocen los recursos de AWS. Así que para una instancia de Amazon EC2, el ID de la instancia podría añadirse a la regla del grupo de seguridad para esa instancia para permitir el tráfico de la instancia. Los clientes también pueden añadir reglas para otros grupos de seguridad, o añadir una regla para el propio grupo de seguridad.  

* Otra gran distinción es que los grupos de seguridad tienen una denegación explícita oculta, lo que significa que se deniega cualquier cosa que no esté explícitamente permitida.

### Prácticas recomendadas adicionales de seguridad de Amazon VPC

* *Utilice múltiples despliegues de Zona de Disponibilidad para tener una alta disponibilidad*. En la siguiente lección aprenderá que los despliegues Multi-AZ se consideran una práctica recomendada para cualquier aplicación que necesite alta disponibilidad. Los recursos adicionales extienden la superficie de ataque potencial y proporcionan una forma rápida de reemplazar los recursos que se han caído debido a la actividad maliciosa.

* *Utilice Amazon CloudWatch para monitorizar sus componentes de Amazon VPC*. CloudWatch no solo sirve para monitorizar sus necesidades de capacidad; también es otra forma de saber si su infraestructura se ha visto comprometida. Detectar y alertar a sus administradores de picos inusuales en las necesidades de CPU, memoria o capacidad de red le ayuda a identificar actividades maliciosas de las que hay que ocuparse lo antes posible. CloudWatch también puede monitorizar sus puntos finales de VPN, gateways NAT y AWS Transit Gateways, todos ellos recursos de Amazon VPC.

* *Utilice los registros de flujo de la VPC para capturar información sobre el tráfico.* Los registros de flujo de VPC capturan información sobre el tráfico IP que entra y sale de las interfaces de red de su Amazon VPC. Son una herramienta excelente para identificar problemas con el tráfico de su red, incluida la actividad no deseada, o incluso para identificar reglas de seguridad que están creando bloqueos innecesarios en el flujo del tráfico deseado. Los datos de registro de flujo no afectan al rendimiento ni a la latencia de la red porque se recopilan fuera de la ruta del tráfico de red. Una vez que haya configurado un log de flujo para su Amazon VPC, puede designar un bucket de Amazon S3 para almacenarlos. Estos logs pueden revisarse manualmente o incluso gestionarse mediante una solución de procesamiento de datos para automatizar la detección de problemas en el tráfico de red. **Nota:** los logs de flujo de la VPC no capturan la carga útil de los paquetes.

* *Encadenamiento de grupos de seguridad.* Encadenar grupos de seguridad añade seguridad por capas. 
    * Permita el acceso del puerto 22 [Secure Shell (SSH)] a cada nivel para la administración.
    * Sólo permita que sus servidores web tengan el puerto 80 (HTTP) o el puerto 443 (HTTPS) abierto a Internet.
    * Sus servidores de aplicaciones sólo permitirían el tráfico originado por el grupo de seguridad del servidor web.
    * Tus servidores de bases de datos sólo permitirán el tráfico originado por el grupo de seguridad del servidor de aplicaciones.

    De esta forma, tus grupos de seguridad están encadenados desde el servidor web al servidor de aplicaciones y a la base de datos, impidiendo accesos no autorizados que no sigan ese patrón.

### Añadir alta disponibilidad

¿Cómo se gestionan las interrupciones en la disponibilidad de los recursos de tu aplicación? ¿Recuerda la Amazon VPC personalizada que se creó en el primer módulo? ¿Cómo puede hacer que esté altamente disponible para reducir las interrupciones y añadir alta disponibilidad y escalabilidad a la Amazon VPC personalizada? 

AWS proporciona balanceadores de carga para lograr alta disponibilidad, tolerancia a fallos y escalado, y también Amazon VPC personalizadas en las que se pueden configurar dos subredes, cada una en una zona de disponibilidad distinta, lo que crea un diseño Multi-AZ.

#### Balanceadores de carga elásticos

Un equilibrador de carga es un recurso utilizado para distribuir las conexiones entrantes entre un grupo de servidores o servicios. Las conexiones entrantes se dirigen al balanceador de carga, que distribuye las conexiones a los servidores o servicios. Los balanceadores de carga son ideales para combinarlos con un grupo de autoescalado de AWS para mejorar la alta disponibilidad, la tolerancia a errores y la escalabilidad de una aplicación.

Elastic Load Balancing (ELB) distribuye automáticamente el tráfico entrante de la aplicación entre varios destinos, como instancias de Amazon EC2, contenedores, direcciones IP, funciones de AWS Lambda y dispositivos virtuales. Puede gestionar la carga variable del tráfico de su aplicación en una única zona de disponibilidad o en varias zonas de disponibilidad. 

Todos los balanceadores de carga elásticos ofrecen alta disponibilidad, escalado automático y la sólida seguridad necesaria para que sus aplicaciones sean tolerantes a fallos. AWS proporciona cuatro tipos de balanceadores de carga; cada uno ofrece ventajas para configuraciones específicas.

#### Balanceadores de carga clásicos

AWS comenzó con un tipo de balanceador de carga que era un balanceador de carga elástico.  El balanceador de carga elástico no ofrecía muchas características, por lo que AWS añadió más características y creó el Application Load Balancer.  A continuación, AWS añadió aún más funciones y lanzó el Network Load Balancer, y así sucesivamente.

El balanceador de carga heredado de AWS, el balanceador de carga elástico, es en realidad el balanceador de carga clásico. El balanceador de carga clásico, el balanceador de carga de aplicaciones, el balanceador de carga de red y el balanceador de carga de gateway conforman la familia de productos conocida como balanceador de carga elástico (ELB).

No se recomienda el uso del Classic Load Balancer a menos que tenga servicios o aplicaciones heredados que necesiten el Classic Load Balancer.  Se recomienda elegir el Application Load Balancer en lugar del Classic Load Balancer siempre que sea posible.

#### Equilibrador de carga de aplicaciones

El Application Load Balancer se conoce como equilibrador de carga de capa 7 del modelo de Interconexión de Sistemas Abiertos (OSI). Capa 7 significa que el Application Load Balancer puede inspeccionar los datos que pasan a través de él y puede entender la capa de aplicación, es decir, HTTP y HTTPs. El Application Load Balancer puede entonces tomar acciones basadas en cosas de ese protocolo como rutas, cabeceras y hosts. 

Todos los balanceadores de carga de AWS son escalables y altamente disponibles. El Application Load Balancer tiene nodos individuales que se ejecutan en cada zona de disponibilidad configurada con el Application Load Balancer. Los balanceadores de carga de aplicaciones pueden estar orientados a Internet o ser internos; la diferencia es que los balanceadores de carga de aplicaciones orientados a Internet tendrán direcciones IP públicas y los balanceadores de carga de aplicaciones internos tendrán direcciones IP privadas. 

Un Application Load Balancer orientado a Internet está diseñado para conectarse desde Internet y esas conexiones del balanceador de carga se conectan contra las instancias de destino. Los balanceadores de carga internos no son accesibles desde Internet y se utilizan para equilibrar cargas dentro de la VPC de Amazon o entre las capas de una aplicación multinivel.

De nuevo, los balanceadores de carga de aplicaciones externos escuchan desde el exterior y envían tráfico a objetivos o grupos de objetivos dentro de una Amazon VPC. Los balanceadores de carga de aplicaciones se facturan a una tarifa por hora y a una tarifa adicional en función de la carga impuesta a su balanceador de carga.

#### Equilibradores de carga de red

Los equilibradores de carga de red tienen ventajas sobre los equilibradores de carga de aplicaciones porque un equilibrador de carga de red no necesita preocuparse del protocolo de capa superior y es mucho más rápido. Los equilibradores de carga de red pueden gestionar cargas de trabajo de alto nivel y escalar a millones de peticiones por segundo. 

Los equilibradores de carga de red pueden asignar direcciones IP estáticas y son más fáciles de integrar con productos de seguridad y cortafuegos. Los balanceadores de carga de red también permiten enrutar solicitudes en varias aplicaciones en una única instancia de Amazon EC2 y admiten el uso de aplicaciones en contenedores.  

Los balanceadores de carga de aplicaciones son excelentes para la compatibilidad con protocolos de capa 7 de gama alta, mientras que los balanceadores de carga de red admiten todos los demás protocolos y pueden manejar millones de solicitudes.

#### Gateway Load Balancers

Los Gateway Load Balancers le permiten desplegar, escalar y gestionar dispositivos virtuales, como cortafuegos, sistemas de detección y prevención de intrusiones y sistemas de inspección profunda de paquetes. Combina una pasarela de red transparente (es decir, un único punto de entrada y salida para todo el tráfico) y distribuye el tráfico a la vez que escala sus dispositivos virtuales con la demanda. 

* Un Gateway Load Balancer opera en la tercera capa del modelo de Interconexión de Sistemas Abiertos (OSI), la capa de red.

* Escucha todos los paquetes IP en todos los puertos y reenvía el tráfico al grupo de destino especificado en la regla de escucha. 

* Mantiene la fidelidad de los flujos a un dispositivo de destino específico utilizando 5-tuple (para flujos TCP/UDP) o 3-tuple (para flujos no TCP/UDP). 

* El Gateway Load Balancer y sus instancias de dispositivo virtual registradas intercambian tráfico de aplicaciones utilizando el protocolo GENEVE en el puerto 6081. Admite un tamaño de unidad de transmisión máxima (MTU) de 8.500 bytes.

Los Gateway Load Balancer utilizan puntos finales del Gateway Load Balancer para intercambiar tráfico de forma segura a través de los límites de la VPC. Un punto final de Gateway Load Balancer es un punto final de VPC que proporciona conectividad privada entre dispositivos virtuales en la VPC del proveedor de servicios y servidores de aplicaciones en la VPC del consumidor de servicios. El Gateway Load Balancer se despliega en la misma VPC que los dispositivos virtuales. Registre los dispositivos virtuales en un grupo de destino para el equilibrador Load Gateway.

* El tráfico hacia y desde un punto final del Gateway Load Balancer se configura utilizando tablas de rutas. 
* El tráfico fluye desde la VPC del consumidor de servicios a través del punto final del Gateway Load Balancer hasta el Gateway Load Balancer en la VPC del proveedor de servicios y, a continuación, vuelve a la VPC del consumidor de servicios. 
* Debe crear el punto final del Gateway Load Balancer y los servidores de aplicaciones en subredes diferentes.
* Esto le permite configurar el punto final del Gateway Load Balancer como el siguiente salto en la tabla de rutas para la subred de la aplicación.

#### Estrategias de diseño de Amazon VPC 

¿Qué es el diseño de su infraestructura? A alto nivel, su diseño de infraestructura es el documento, diagrama o visión que abarca todas las características de su infraestructura que soporta sus aplicaciones hospedadas. Un diseño incluirá los requisitos, las restricciones y las suposiciones que abordan la disponibilidad, la capacidad de gestión, el rendimiento, la capacidad de recuperación y la seguridad de los sistemas y recursos que soportan y alojan sus flujos de trabajo de automatización empresarial. Para esta sección, el enfoque se limitará a una discusión del diseño de la arquitectura de aplicaciones de un solo nivel y de varios niveles.

Una estrategia eficaz para proteger los recursos y los datos confidenciales en la Nube requiere una buena comprensión de los patrones generales de seguridad de los datos y una clara asignación de estos patrones a los controles de seguridad de la Nube. Estos controles pueden aplicarse en un diseño de aplicación de varios niveles, a detalles de nivel de implementación específicos de los almacenes de datos, a la capa de base de datos donde residen Amazon Relational Database Service (Amazon RDS) y Amazon DynamoDB o a la capa web o de aplicación donde residen las instancias de Amazon EC2. De nuevo, ¿cuál es su diseño de infraestructura?  

AWS proporciona el Marco de Adopción de la Nube de AWS (AWS CAF) que ofrece orientación y mejores prácticas para ayudar a construir un enfoque integral de la computación en la nube en toda su organización. Dentro de este marco, la perspectiva de seguridad del AWS CAF cubre cinco capacidades clave:

* AWS Identity and Access Management (IAM): Defina, aplique y audite los permisos de usuario en todos los servicios, acciones y recursos de AWS.

* Control de detecciones: Mejore su postura de seguridad, reduzca el perfil de riesgo de su entorno y obtenga la visibilidad que necesita para detectar problemas antes de que afecten a su negocio.

* Seguridad de la infraestructura: Reduzca la superficie de la infraestructura que administra y aumente la privacidad y el control de su infraestructura general en AWS.

* Protección de datos: Implemente las salvaguardas adecuadas que ayuden a proteger los datos en tránsito y en reposo mediante el uso de servicios cifrados integrados de forma nativa.

* Respuesta a incidentes: Definir y poner en marcha una respuesta a incidentes de seguridad como guía para la planificación de la seguridad.

El primer paso a la hora de implementar la perspectiva de seguridad de AWS CAF es pensar en la seguridad desde la perspectiva de los datos. Es imprescindible conocer y comprender las necesidades empresariales, los objetivos empresariales y la estructura de la cuenta desde el principio. Los clientes deben incluir esto desde el principio en el proceso de diseño, ya que esto impulsará la estrategia de la cuenta y el diseño de la cuenta de vuelta a la Estructura Bien Arquitecturada de AWS. En lugar de pensar en el diseño on-premises y off-premises y en la seguridad de los datos, piense en los datos que está protegiendo, cómo se almacenan y quién tiene acceso a ellos.

#### Buenas prácticas: Limitar la exposición de los recursos

Cuando construya su Amazon VPC, piense en cuántas capas de seguridad va a poner entre sus posibles atacantes y sus recursos críticos.

**Arquitectura de aplicación de un solo nivel**

Una arquitectura de aplicación de un solo nivel pone todo en una subred. Esto expone todos sus recursos a cualquier atacante que consiga acceder a su red. Utilizando subredes, su arquitectura de red puede proporcionar capas adicionales de seguridad.

### ¿Debe utilizar siempre la arquitectura de varios niveles?

Puede haber situaciones en las que su Amazon VPC no necesite tres niveles. Puede que ni siquiera necesite dos niveles. Puede que incluso tenga que considerar la posibilidad de conectar varias Amazon VPC juntas en lugar de colocar todo en una Amazon VPC. Veamos brevemente las circunstancias en las que podría considerar cada uno de estos tipos de diseños.

Si solo intenta ejecutar una aplicación sencilla, como un blog personal o un sitio web, puede que no resulte rentable crear arquitecturas complejas de varios niveles. 

Debería considerar una Amazon VPC de un solo nivel si su aplicación:

* No utiliza datos privados

* Puede no estar disponible durante periodos prolongados si algo falla (considere también los despliegues Single-AZ)

* Solo la utilizará usted

Aunque aquí le hemos proporcionado un patrón de diseño común en la Amazon VPC de tres niveles, también podría crearla con dos niveles o más de tres niveles, en función de la arquitectura de su aplicación. A este patrón lo denominamos diseño de Amazon VPC de n niveles.

En última instancia, debe pensar en cuántos niveles necesita para limitar el acceso y el acoplamiento estrecho en la medida de lo posible. Por eso recomendamos, por ejemplo, separar la capa de datos de la capa de aplicación.

Una buena práctica consiste en enrutar juntas sólo las cosas que deban enrutarse juntas. Además, utiliza grupos de seguridad y ACLs de red para limitar el tráfico enrutado tanto como sea posible. Recuerda:

* Las subredes y Amazon VPC no cuestan dinero, sólo cuestan direcciones IP.

* Cada subred que crees tiene que reservar cinco direcciones IP para AWS, las cuatro primeras y la última.

* No se puede cambiar el tamaño de un bloque CIDR una vez creado, por lo que cualquier dirección IP que reserves en ese bloque CIDR puede quedarse ahí hasta que elimines esa subred. 

* Almacenar recursos en subredes separadas aumenta las capas de seguridad entre ellas, pero demasiadas subredes pueden hacer que su Amazon VPC se quede sin direcciones IP.

Existen muchas otras maneras de construir su arquitectura de red de AWS más allá de una Amazon VPC.

* La interconexión de VPC le permite conectar dos Amazon VPC sin tener que atravesar la Internet pública. Puede conectar Amazon VPC en su propia cuenta o en cuentas diferentes, en una región o entre regiones.

* Las opciones de VPN de Amazon VPC y AWS Direct Connect pueden conectar su infraestructura on-premise con sus Amazon VPC y otros recursos de AWS a través de conexiones privadas.
* Los puntos de enlace de VPC son conexiones privadas entre su Amazon VPC y otro servicio de AWS que no reside en su Amazon VPC. Estas conexiones no requieren acceso a Internet.

* Esto se trata con más detalle en el curso Opciones de conectividad de red de AWS.

### Algunas consideracines adicionales

* Al diseñar su instancia de Amazon VPC, debe tener en cuenta el número de direcciones IP necesarias y el tipo de conectividad con el centro de datos antes de elegir el bloque CIDR. El tamaño admisible del bloque oscila entre una máscara de red /16 y una máscara de red /28.

* No puede alterar ni modificar el bloque CIDR de una Amazon VPC desplegada, por lo que es mejor elegir un bloque CIDR que tenga más direcciones IP de las necesarias. Al diseñar su arquitectura de Amazon VPC para comunicarse con el centro de datos on-premise, el rango CIDR utilizado en Amazon VPC no debe solaparse ni provocar un conflicto con el bloque CIDR del centro de datos on-premise.

* Recomendamos utilizar Amazon VPC independientes para los entornos de desarrollo, producción y ensayo. Si decide mantener todos esos entornos en una Amazon VPC, asegúrese de proporcionar el máximo aislamiento posible mediante ACL de red, subredes y otros recursos de Amazon VPC.
